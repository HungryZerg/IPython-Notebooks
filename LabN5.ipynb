```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

### Задача

Дано  3  нечетких  множества  $A,  B,  C$  (заданы  их  функции
принадлежности).  Построить  функцию  принадлежности  нечеткого
множества $D = A \cup \overline{B} \cap C$ и определить степень принадлежности одного
элемента множеству $D$, используя алгебраический способ.

```python
fig, axs = plt.subplots(1,3, figsize=(18,4))
axis_font = {'fontname':'Arial', 'size':'16'}
axs[0].axis([0,15,0,1.5])
axs[0].set_ylabel("$\mu_A$", **axis_font)
axs[0].set_xlabel("X")
axs[0].set_xticks([0,2,8,13])
axs[0].set_yticks([0,0.5, 1])
axs[0].plot([0,2,8,13, 20], [0,0,0.5,0, 0], 'r-',linewidth=2)

axs[1].axis([0,15,0,1.5])
axs[1].set_ylabel("$\mu_B$", **axis_font)
axs[1].set_xlabel("X")
axs[1].set_xticks([0,2,8,9,14])
axs[1].set_yticks([0,0.25, 0.5, 1])
axs[1].plot([0,2,8,9,14, 20], [0,0,0.5,0.25,0, 0], 'r-',linewidth=2)

axs[2].axis([0,15,0,1.5])
axs[2].set_ylabel("$\mu_C$", **axis_font)
axs[2].set_xlabel("X")
axs[2].set_xticks([0,4,5,7,8])
axs[2].set_yticks([0, 0.5, 1])
axs[2].plot([0,4,5,7,8,20], [0,0,1,1,0,0], 'r-',linewidth=2)
```

### Описание процесса  решения.

Для  построения  функции
принадлежности нового множества необходимо:
1.  Определить последовательность выполнения операций в формуле.
2. Построить  на  отдельных  графиках  промежуточные  множества, согласно  определенной  последовательности  действий. 
Свести промежуточные множества на одном графике и определить итоговую функцию принадлежности.
3. Используя определенный в задаче метод, определить аналитически степень принадлежности элемента,
входящего в ядро итогового множества.
4. Проверить  аналитические вычисления по построенному графику функции принадлежности.

### Решение.

- 1) Множество $D = A \cup \overline{B} \cap C$, значит, последовательность операций будет следующей:  $\overline{B}$,  $\overline{B} \cap C$,  $A \cup \overline{B} \cap C$.
- 2) Построим  согласно  этой  последовательности  операций  графики функций принадлежности:

```python
fig, axs = plt.subplots(1,3, figsize=(18,4))

axs[0].axis([0,15,0,1.5])
axs[0].set_ylabel("$\mu_{\overline{B}}$", **axis_font)
axs[0].set_xlabel("X")
axs[0].set_xticks([0,2,8,9,14])
axs[0].set_yticks([0, 0.25, 0.5, 1])
axs[0].plot([0,2,8,9,14, 20], [1,1,0.5,0.75,1, 1], 'r-',linewidth=2)

axs[1].axis([0,15,0,1.5])
axs[1].set_ylabel("$\mu_{\overline{B} \cap C}$", **axis_font)
axs[1].set_xlabel("X")
axs[1].set_xticks([0,4,5,7,8])
axs[1].set_yticks([0,0.25, 0.5, 1])
axs[1].plot([0,2,4,4.77,7.45,8, 20], [0,0,0,0.77,0.55, 0, 0], 'r-',linewidth=2)

axs[2].axis([0,15,0,1.5])
axs[2].set_ylabel("$\mu_{A \cup \overline{B} \cap C}$", **axis_font)
axs[2].set_xlabel("X")
axs[2].set_xticks([0,2,4,5,7,8,13])
axs[2].set_yticks([0,0.25, 0.5, 0.75, 1])
axs[2].plot([0,2,4.18,4.77,7.45,7.54,8,13, 20], [0,0,0.18,0.77,0.55,0.46,0.5,0, 0], 'r-',linewidth=2)
```

- 3) Ядро множества $D$ состоит из элементов из интервала $(2,13)$. Выберем элемент $8$.

$\mu_A (8) = 0.5$

$\mu_B (8) = 0.5$

$\mu_C (8) = 0$

$\mu_\overline{B}(8) = 1-\mu_\overline{B}(8) = 1-0.5 = 0.5$

$\mu_{\overline{B} \cap C}(8) = \mu_\overline{B}(8)*\mu_C(8) = 0.5 * 0 = 0$

$\mu_{A \cup \overline{B} \cap C}(8) = \mu_A (8) + \mu_{\overline{B} \cap C}(8) - \mu_A (8)*\mu_{\overline{B} \cap C}(8) = 0.5 + 0 - 0.5*0 = 0.5$

- 4) $\mu_D (8) = 0.5$

```python
plt.figure(figsize=(6,4))
plt.axis([0,15,0,1.5])
plt.ylabel("$\mu_D$", **axis_font)
plt.xlabel("X")
plt.xticks([0,2,4,5,7,8,13])
plt.yticks([0,0.25, 0.5, 0.75, 1])
plt.plot([0,8],[0.5, 0.5], color='blue', ls='dashed')
plt.plot([8,8],[0,0.5], color='blue', ls='dashed')
plt.plot([0,2,4.18,4.77,7.45,7.54,8,13, 20], [0,0,0.18,0.77,0.55,0.46,0.5,0, 0], 'r-')
```
